```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
df["genres"] = df["genres"].apply(convert)
```

```python
df.head(3)
```

```python
df['keywords'][0]
```

```python
df["keywords"] = df["keywords"].apply(convert)
```

```python
df["cast"][0]
#extract the name of the actor
```

```python
def convert2(obj) :
  L = []
  counter = 0
  for i in ast.literal_eval(obj) :
    if counter != 3 :
      L.append(i["name"])
      counter += 1
    else :
      break
  return L
```

```python
df["cast"] = df["cast"].apply(convert2)
```

```python
df.head(5)
```

```python
df["crew"][0]
# job -> director
```

```python
def extract_director (obj) :
  L = []
  for i in ast.literal_eval(obj) :
    if i ["job"] == 'Director' :
      L.append(i["name"])
      break
  return L
```

```python
df["crew"] = df["crew"].apply(extract_director)
```

```python
df["overview"] = df["overview"].apply(lambda x:x.split())
```

```python
df.head()
```

```python
df['genres']= df['genres'].apply(lambda x: [i. replace (" ","") for i in x ])
df['keywords']= df['keywords'].apply(lambda x : [ i.replace(" ", "") for i in x])
df['cast']=df['cast'].apply(lambda x : [i.replace (" ", "") for i in x])
df['crew'] = df['crew'].apply(lambda x : [i.replace (" ","") for i in x])
df['keywords'] = df['keywords'].apply (lambda x : [i.replace (" ", "") for i in x ])
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast']
```

```python
df2 = df[['movie_id', 'title', 'tags']]
```

```python
df2.head(1)
```

```python
df2['tags'] = df2['tags'].apply(lambda x : " ".join(x))
```

```python
df2.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer ()
```

```python
def stem (text) :
  y = []

  for i in text.split () :
    y.append(ps.stem(i))
  return " ".join (y)
```

```python
df2['tags'] = df2['tags'].apply(stem)
```

```python
df2['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer (max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df2['tags']).toarray()
```

```python
 from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
def recommend (df) :
  df_index = df2[df2['title'] == 'Batman Begins'].index[0]
  distances = similarity[df_index]
  movie_list = sorted(list(enumerate(distances)), reverse = True, key= lambda x : x[1]) [1:6]

  for i in movie_list :
    print (df2.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
